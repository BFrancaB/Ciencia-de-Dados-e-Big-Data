## Exercicio 6

## Exercicio 7
Crie um subconjunto de dados com as colunas:

Respondent, Country, Student, Employment, FormalEducation, UndergradMajor, Exercise, Age

```python
df_conjunto = df_brasil[['Respondent', 'Country', 'Student', 'Employment', 'FormalEducation', 'UndergradMajor', 'Exercise', 'Age']]

df_conjunto
```

## Exercicio 8
Exiba as contagens de frequência dos valores únicos das colunas: Student, Exercise, Employment, FormalEducation, UndergradMajor e Age

```python
print('Student:')
print(df_conjunto['Student'].value_counts())
print('\nExercise: ')
print(df_conjunto['Exercise'].value_counts())
print('\nEmployment: ')
print(df_conjunto['Employment'].value_counts())
print('\nFormalEducation: ')
print(df_conjunto['FormalEducation'].value_counts())
print('\nUndergradMajor: ')
print(df_conjunto['UndergradMajor'].value_counts())
print('\nAge: ')
print(df_conjunto['Age'].value_counts())
print('\n')
df_conjunto.describe(include='all')
```
